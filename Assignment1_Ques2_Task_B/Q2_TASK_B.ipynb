#**Importing Necessary Libraries**

```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
```

#**Generating the Spectrograms**

```python

def generate_spectrogram(audio_path, title, output_file):
    y, sr = librosa.load(audio_path, sr=None)
    D = librosa.amplitude_to_db(np.abs(librosa.stft(y)), ref=np.max)

    plt.figure(figsize=(10, 5))
    librosa.display.specshow(D, sr=sr, x_axis='time', y_axis='log')
    plt.colorbar(format='%+2.0f dB')
    plt.title(title)
    plt.savefig(output_file)
    plt.close()

songs = {
    "Hoshwalon Ko Khabar Kya": "/content/sample_data/hoshwalon_ko_khabar_kya.mp3",
    "Tum Hi Ho": "/content/sample_data/tum_hi_ho.mp3",
    "Kesariya Balam": "/content/sample_data/kesariya_balam.mp3",
    "Om Jai Jagdish Hare": "/content/sample_data/om_jai_jagdish.mp3"
}

for title, file_path in songs.items():
    output_file = f"{title.replace(' ', '_')}_spectrogram.png"
    generate_spectrogram(file_path, title, output_file)
    print(f"Saved spectrogram: {output_file}")
```

#**Plotting the Spectrograms**
